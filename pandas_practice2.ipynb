```python
import pandas as pd
#import os
```

```python
student_sv_file_to_read = "resources/clean_students_complete.csv"
school_csv_file_to_read = "resources/schools_complete.csv"

#student_data_to_load = os.path.join("Resources", "students_complete.csv")
```

```python
students_df = pd.read_csv(student_sv_file_to_read)
school_df = pd.read_csv(school_csv_file_to_read)
```

```python
school_data_complete_df = pd.merge(students_df, school_df, on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
remove_cols = [col for col in school_data_complete_df.columns if 'Unnamed' in col]
school_data_complete_df.drop(remove_cols, axis='columns', inplace=True)
```
